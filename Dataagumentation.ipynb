```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array,load_img
```

```python
input_folder = "/content/drive/MyDrive/dog"
output_folder = "/content/drive/MyDrive/Colab Notebooks"
os.makedirs(output_folder, exist_ok=True)
```

```python
datagen = ImageDataGenerator(
rotation_range=30,
width_shift_range=0.1,
height_shift_range=0.1,
zoom_range=0.2,
shear_range=0.2,
horizontal_flip=True,
brightness_range=[0.7, 1.3],
fill_mode='nearest'
)
```

```python
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith((".jpg","png","jpeg"))]
print("Found images:", len(image_files))
count = 0
for img_name in image_files:
  img_path = os.path.join(input_folder, img_name)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image = load_img(img_path)
image_arr = img_to_array(image)
image_arr = image_arr.reshape((1,) + image_arr.shape)
```

```python
i = 0
for batch in datagen.flow(image_arr, batch_size=1,
                               save_to_dir=output_folder,
                               save_prefix=f"aug_{count}",
                               save_format="jpg"):
  i += 1
  if i >= 50:
     break
count += 1

print("Data Augmentation Completed!")
print("Augmented images saved in:", output_folder)
```
